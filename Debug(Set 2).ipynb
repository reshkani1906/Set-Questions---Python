```python
#2.Find the error in the following Python code:
def my_function():
    x += 5
    print(x)

x = 10
my_function()
```

```python
#ANSWER
#Error: x += 5 tries to change x inside the function, but Python thinks x is a local variable and it has no value yet.

def my_function():
    global x
    x += 5
    print(x)

x = 10
my_function()
```
