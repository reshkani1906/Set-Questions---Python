```python
#4.Question:
What will happen when you run this code?

a = 10
b = 0
result = a / b
print(result)
```

```python
#Answer

a = 10
b = 0

if b != 0:
    result = a / b
    print(result)
else:
    print("Error: Division by zero is not allowed.")
```

```python
#5.What error will occur in the following code?

num = int("xyz")
print(num)
```

```python
#Answer

user_input = "xyz"

if user_input.isdigit():
    num = int(user_input)
    print(num)
else:
    print(f"Error: '{user_input}' contains non-numeric characters.")
```

```python
#6.Identify the issue in this code:

data = {"name": "John", "age": 30}
print(data["city"])
```

```python
#Answer

data = {"name": "John", "age": 30}
city = data.get("city", "Not Specified")

print(city)
```

```python
#7.What will happen when you run this code?
num = 1
while num < 5:
    print(num)
```

```python
#Answer

num = 1
while num < 5:
    print(num)
    num += 1
```
